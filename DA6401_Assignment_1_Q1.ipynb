```python
import wandb
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import fashion_mnist
import wandb
wandb.init(project="DA6401_Assignement_1_CE24D904")
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
x_train = x_train.reshape(x_train.shape[0], 784)/256   
x_test = x_test.reshape(x_test.shape[0], 784)/256  
print("X_train shape:", x_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", x_test.shape)
print("y_test shape:", y_test.shape)
```

Question-1 Download the fashion-MNIST dataset and plot 1 sample image for each class as shown in the grid below. Use from keras.datasets import fashion_mnist for getting the fashion mnist dataset.

```python
import matplotlib.pyplot as plt

# Enable inline plotting for Jupyter Notebooks
%matplotlib inline

# Define the class labels
class_labels = [
    "T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
    "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"
]

# Assume x_train and y_train are defined and populated with the Fashion-MNIST data
samples = {}
for img, label in zip(x_train, y_train):
    if label not in samples:
        samples[label] = img
    if len(samples) == 10:
        break

# Create a figure with subplots
fig, axes = plt.subplots(2, 5, figsize=(10, 5))

# Plot each sample
for ax, (label, img) in zip(axes.ravel(), samples.items()):
    img_reshaped = img.reshape(28, 28)  # Ensure the image is 28x28
    ax.imshow(img_reshaped, cmap='gray')
    ax.set_title(class_labels[label])
    ax.axis('off')

plt.tight_layout()
plt.show()

```
